```python
x_jk = np.arange(0, 57, 5) #range
x_ch = np.arange(0, 2800, 100) #range
x_ku = np.arange(0, 170, 20) #range
x_pht = np.arange(0, 12.1, 2) #range
x_km = np.arange(0, 60, 20) #range
x_km = np.arange(0, 60, 20) #range
x_to = np.arange(0, 1700, 100) #range

# Generate grafik
jk_sebentar = fuzz.trapmf(x_jk, [0,25, 32, 57])
jk_curahhujan = fuzz.trapmf(x_ch, [0,300, 2500, 2800])
jk_kelembapanudara = fuzz.trapmf(x_ku, [0,80, 90, 170])
jk_phtanah = fuzz.trapmf(x_pht, [0,5.6, 6.5, 12.1])
jk_km = fuzz.trapmf(x_km, [0,0,30,60])
jk_to = fuzz.trapmf(x_to, [0,700, 1000, 1700])

# Visualize
fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(nrows=6, figsize=(12, 6))

ax1.plot(x_jk, jk_sebentar, 'b', linewidth=1.5, label='Temperatur')
ax1.set_title('Bawang Merah')
ax1.legend()
ax2.plot(x_ch, jk_curahhujan, 'r', linewidth=1.5, label='Curah Hujan')
ax2.set_title('Curah Hujan')
ax2.legend()
ax3.plot(x_ku, jk_kelembapanudara, 'b', linewidth=1.5, label='Kelembapan Udara')
ax3.set_title('Kelembapan Udara')
ax3.legend()
ax4.plot(x_pht, jk_phtanah, 'r', linewidth=1.5, label='Ph Tanah')
ax4.set_title('Ph Tanah')
ax4.legend()
ax5.plot(x_km, jk_km, 'b', linewidth=1.5, label='Kemiringan Tanah')
ax5.set_title('Kemiringan Tanah')
ax5.legend()
ax6.plot(x_to, jk_to, 'r', linewidth=1.5, label='Topografi')
ax6.set_title('Topografi')
ax6.legend()
plt.tight_layout()
```

```python
level_jk_seb = fuzz.interp_membership(x_jk, jk_sebentar, 26)
level_jk_seb
```

```python
data["interpolasi BM: temperatur"] = fuzz.interp_membership(x_jk, jk_sebentar, data["C"])*5
data["interpolasi BM: curah hujan"] = fuzz.interp_membership(x_ch, jk_curahhujan, data["D"])*5
data["interpolasi KU: kelembapan udara"] = fuzz.interp_membership(x_ku, jk_kelembapanudara, data["E"])*5
data["interpolasi PT: Ph Tanah"] = fuzz.interp_membership(x_pht, jk_phtanah, data["H"])*5
data["interpolasi KT: Kemiringan Tanah"] = fuzz.interp_membership(x_km, jk_km, data["I"])*5
data["interpolasi T: Topografi"] = fuzz.interp_membership(x_to, jk_to, data["J"])*5
data
```
